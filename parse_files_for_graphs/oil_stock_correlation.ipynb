```python
import pandas as pd
df = pd.read_csv("data/commodity.csv")
df.head()
```

```python
wti_df = df[df["Commodity And Exchange"] == "WTI CRUDE OIL (DOLLARS PER BARREL)"] 
# nymex_df = df[df["Commodity And Exchange"] == "NYMEX CRUDE OIL FUTURES"]
wti_df.head()
```

```python
wti_df = wti_df.iloc[:, [0,1]]
wti_df.head()
```

```python
wti_df = wti_df.sort_values(by="Date")
wti_df.head()
```

```python
df = pd.read_csv("data/global_stock_exchange.csv")
df.head()
```

```python
companies = df["Stock Name"].unique()
dfs = []
for c in companies:
#     print(c)
    dfs.append(df[df["Stock Name"] == c])

dfs[0].head()
```

```python
for i in range(len(dfs)):
    dfs[i] = dfs[i].loc[:, ["Value", "Date"]]

dfs[0].head()
```

```python
merge_dfs = []
for df in dfs:
    merge_dfs.append(wti_df.merge(df, on="Date"))

merge_dfs[0]
```

```python
for merge_df in merge_dfs:
    merge_df.set_index("Date")
    merge_df.plot()
```

```python

for merge_df in merge_dfs:
    oil_price = merge_df["Value_x"]
    stock_price = merge_df["Value_y"]
    print(oil_price.corr(stock_price))
```
