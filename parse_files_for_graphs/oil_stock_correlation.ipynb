```python
import pandas as pd
df = pd.read_csv("data/commodity.csv")
df.head()
```

```python
wti_df = df[df["Commodity And Exchange"] == "WTI CRUDE OIL (DOLLARS PER BARREL)"] 
# nymex_df = df[df["Commodity And Exchange"] == "NYMEX CRUDE OIL FUTURES"]
wti_df.head()
```

```python
wti_df = wti_df.iloc[:, [0,1]]
wti_df.head()
```

```python
wti_df = wti_df.sort_values(by="Date")
wti_df.head()
```

```python
df = pd.read_csv("data/global_stock_exchange.csv")
df.head()
```

```python
companies = df["Stock Name"].unique()
dfs = []
for c in companies:
#     print(c)
    dfs.append(df[df["Stock Name"] == c])

dfs[0].head()
```

```python
for i in range(len(dfs)):
    dfs[i] = dfs[i].loc[:, ["Value", "Date"]]

dfs[0].head()
```

```python
merge_dfs = []
for df in dfs:
    merge_dfs.append(wti_df.merge(df, on="Date"))
# y is stock price. (verified)
merge_dfs[0].set_index('Date').loc['2020-07-08':'2020-08-08', :]
```

```python
for merge_df in merge_dfs:
    merge_df = merge_df.set_index("Date")
    merge_df.plot()
```

```python
import matplotlib.pyplot as plt
for merge_df in merge_dfs:
    oil_price = merge_df["Value_x"]
    stock_price = merge_df["Value_y"]
    lag = 10
    n = len(oil_price)
    oil_price_l = oil_price[lag:]
    stock_price_l = stock_price[:n-lag]
    
    oil_price_l = oil_price_l.reset_index()
    stock_price_l = stock_price_l.reset_index()
    
    del oil_price_l['index']
    del stock_price_l['index']
    
#     print(oil_price_l)
#     print(stock_price_l)
    plt.figure()
    l2,  = plt.plot(oil_price_l)
    l1,  = plt.plot(stock_price_l)
    plt.legend([l1, l2], ["stock", "oil"])
```

```python
merge_df = merge_dfs[1]
oil_price = merge_df["Value_x"]
stock_price = merge_df["Value_y"]
```

```python
def find_lagged_correlation(series1, series2, lag):
    assert len(series1) == len(series2)
    n = len(series1)
    s1 = series1[:n-lag]
    s2 = series2[lag:n]
    assert len(s1) == len(s2)
    return s1.corr(s2)
```

```python
print_stats = False
plot_stats = True
c = 1
for merge_df in merge_dfs:
    oil_price = merge_df["Value_x"]
    stock_price = merge_df["Value_y"]
    print(len(oil_price))
    print_stats and print("df: {}".format(c))
    print_stats and  print()
    c += 1 
    corrs = []
    for lag in range(140):
        corr = find_lagged_correlation(stock_price, oil_price, lag)
        corrs.append(corr)
        print_stats and print('lag: {}, corr: {}'.format(lag, corr))
    
    print_stats and print()
    plot_stats and plt.figure()
    plot_stats and plt.plot(corrs)
```
